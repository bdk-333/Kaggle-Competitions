### Analysis on Numeric cols:

```python
def show_histogram(col):
    sns.histplot(train_df, x=train_df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
def show_boxplot(col):
    sns.boxplot(train_df, x=train_df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
def get_skewness(col):
    print('-'*10, train_df[col].skew(), '-'*10)
    print()
```

- Age is uniform, no skewness and no outliers, can use mean to impute.
- Annual Income is heavely right skewed, so a lot of outliers, median is better than mean for imputation.
- Health Score is close to normal, just a little right-skewed, no outliers, can use median to impute.
- Previous claims has some outliers, use mean for imputation.
- Vehicle age is uniform, no outliers, can use mean for imputation.
- Credit card score is a little skewed, no outliers, can use median to impute.
- Premium Amount is also heavily skewed to the right, thus has a lot of outliers.
- Policy Start date has uniform distribution for years, months and days.

### Imputation transformer for filling missing values

```python
for i, col in enumerate(train_df.columns):
    print(f'{col} index is: {i}')
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
imputer = ColumnTransformer([
    ('mean_imputer', SimpleImputer(strategy='mean'), [0, 4, 10, 11, 13]), # Age, no. of dependents, previous claims, vehicle age, insurance duration
    ('median_imputer', SimpleImputer(strategy='median'), [2, 7, 12]), # annual income, health score, credit card
    ('most_frequent_imputer', SimpleImputer(strategy='most_frequent'), [3, 6, 14]) # marital status, occupation, customer feedback
], remainder='passthrough')
```
